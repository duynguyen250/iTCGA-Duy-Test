```python
import fastq as f
```

```python
data = f.read("LIB049253_GEN00202129_S12_L004_R1_copy.fastq")
```

```python
data = list(data)
```

```python
data[0]
```

```python
data[0].info
```

```python
qual = 0
for n,c in enumerate(data[0].qstr):
    qual += (ord(c) -33)
print(qual/n)
```

```python
for i in range(10):
    print(f'line {i}:')
    for key in ['qual','qual_variance']:
        if key == 'qual_variance':
            var = data[i].info[key]**(1/2)
            print(f' {key}: {var}')
        else:
            print(f' {key}: {data[i].info[key]}')
    print('\n')
```
